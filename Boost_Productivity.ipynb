```python
import pandas as pd 
import time 

data_URL = "https://raw.githubusercontent.com/keitazoumana/Experimentation-Data/main/vgsales.csv" 

game_df = pd.read_csv(data_URL)

game_df.head()
```

## For Loop Vs. Vectorization

```python
begin = time.time()

for index, row in game_df.iterrows():

    game_df.loc[index, 'Sum_NA_EU'] = row["NA_Sales"] + row["EU_Sales"]

print(time.time() - begin)
```

```python
game_df.head()
```

```python
begin = time.time()

game_df["Sum_NA_EU"] = game_df["NA_Sales"] + game_df["EU_Sales"] 

print(time.time() - begin)
```
